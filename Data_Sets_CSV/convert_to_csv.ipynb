```python
xl = pd.ExcelFile('./Appendices/data-table-B12.XLS')

print(xl.sheet_names)  # see all sheet names

# xl.parse(sheet_name)  # read a specific sheet to DataFrame
```

```python
for root, dir, files in os.walk("."):
    for f in files:
        if len(f) >= 4:
            if f[-4:].lower() == '.xls':
                f_path = root + '/' + f
                xl = pd.ExcelFile(f_path)
                for sheet in xl.sheet_names:
                    print(sheet)
                    df = xl.parse(sheet)
                    df.to_csv(f_path[:-4] + ' - ' + sheet + '.csv', index = False)
                    # os.remove(f_path)
```

```python
for root, dir, files in os.walk("."):
    for f in files:
        if len(f) >= 4:
            if f[-4:].lower() == '.xls':
                f_path = root + '/' + f
                os.remove(f_path)
```
